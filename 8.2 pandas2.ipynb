```python
df_sales = pd.read_csv(r'C:\Users\Manoj\OneDrive\Desktop\data\sales_data_final.csv')
```

```python
df_sales
```

```python
#using anonymous or lambda function to add new coloumn or find postive and negative value of profit
```

```python
def profit_flag(a) :
    if a >=0 :
        return 'positive'
    else :
        return 'negative'
```

```python
profit_flag(-45)
```

```python
profit_flag(45)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales
```

```python
#add new coloumn as lenght of customer name
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales
```

```python
#anonymous or lambda function(creat function without name)  
```

```python
def quantity_flag(a) :
    if a<10 :
        return 'low'
    elif a>10 and a<20 :
        return 'medium'
    else :
        return 'high'    
    
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(20)
```

```python
# multleply value in new coloumn
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a*2)
```
