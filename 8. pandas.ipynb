# Series

```python
#series give only one dimension data
```

```python
import pandas as pd
```

```python
x = [2,3,5,6]
i = pd.Series(x)  #it will iterate the data with index no.
print(i)
print(type(i))
print(i[2])    
```

```python
x = [2,3,5,6]
i = pd.Series(x,index=['a','b','c','d'])      #we can also change the index
print(i)
```

```python
x = [2,3,5,6]
i = pd.Series(x,index=['a','b','c','d'],dtype='float')      #we can also change data type
print(i)
```

```python
x = [2,3,5,6]
i = pd.Series(x,index=['a','b','c','d'],dtype='float',name='python')      #we can also gibe the data name
print(i)
```

```python
y = [9,2,1,9,5,3,8,2,5,6]
```

```python
pd.value_counts(y)
```

```python
#also pass tuple, dict, list, set , str, int etc.
```

```python
y = (9,2,1,9,5)
pd.Series(y)
```

```python
y = {9 : [2,1,9,5],
     3 : [8,2,5,6]}
pd.Series(y)
```

```python
s = pd.Series(11)
print(s)
```

```python
s = pd.Series(11,index=[1,2,3,4,5,6,7])   #if give additional index then value will be reapted
print(s) 
```

```python
s1 = pd.Series(11,index=[1,2,3,4,5,6,7])   #can add different index value and give blank as nan
s2 = pd.Series(11,index=[1,2,3,4])
print(s1+s2)
```

# Arithmetic operation in pandas

```python
var4 = pd.DataFrame({'x':[2,5,6,8],'y':[5,8,9,7]})
var4
```

```python
#addition of columns
var4['z'] = var4['x'] + var4['y']
var4
```

```python
#minus of columns
var4['z'] = var4['x'] - var4['y']
var4
```

```python
#multiply of columns
var4['z'] = var4['x'] * var4['y']
var4
```

```python
#devide of columns
var4['z'] = var4['x'] / var4['y']
var4
```

```python
var5 = pd.DataFrame({'x':[20,50,60,80],'y':[5,8,9,7]})
var5
```

```python
var5['python'] = var5['x']>=60
var5['python1'] = var5['y']<=7
```
